# 1. 인구 피라미드

```python
import pandas as pd
```

## 남자 데이터 정의

```python
df_m = pd.read_excel('201108_201108_연령별인구현황_월간.xlsx', skiprows = 3, index_col = '행정기관', usecols = 'B, E:Y')
df_m
```

```python
df_m.iloc[0] = df_m.iloc[0].str.replace(',', '').astype(int) # 1,195,951 -> 1195951 (정수형)
```

## 여자 데이터 정의

```python
df_w = pd.read_excel('201108_201108_연령별인구현황_월간.xlsx', skiprows = 3, index_col = '행정기관', usecols = 'B, AB:AV')
df_w
```

```python
df_m.columns
```

```python
df_w.columns
```

```python
df_w.columns = df_m.columns # 컬럼명 통일
df_w.columns
```

```python
df_w.iloc[0] = df_w.iloc[0].str.replace(',', '').astype(int) # 1,195,951 -> 1195951 (정수형)
```

## 데이터 시각화

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # 글씨체
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상을 해결
```

```python
plt.figure(figsize = (10, 7))
plt.barh(df_m.columns, -df_m.iloc[0] // 1000) # 단위 : 천 명 # 여자와 방향 반대 위해 (-)
plt.barh(df_w.columns, df_w.iloc[0] // 1000) # 단위 : 천 명
plt.title('2011 대한민국 인구 피라미드')
plt.savefig('2011_인구피라미드.png', dpi = 100)
plt.show()
```
