```python
x = np.linspace(0,5,10)
y = x**2
plt.plot(x,y);
```

```python
x = np.linspace(0,10,20)
y = x**2.0
plt.plot(x,y,'bo-',linewidth=3,markersize=5);
```

```python
plt.plot(x,y,'gs-',linewidth=1,markersize=3);
```

```python
x = np.linspace(0,10,20)
y1 = x**2.0
y2 = x**1.5
plt.plot(x,y1,'bo-',linewidth=2,markersize=12,label='First')
plt.plot(x,y2,'gs-',linewidth=2,markersize=12,label='Second')
plt.xlabel("X")
plt.ylabel("Y")
plt.axis([-0.5,10.5,-5,105])
plt.legend(loc=2)
plt.savefig("mplot.pdf");
```

```python
x = np.logspace(-1,1,20)
y1 = x**2.0
y2 = x**1.5
plt.plot(x,y1,'bo-',linewidth=2,markersize=5,label='F')
plt.plot(x,y2,'gs-',linewidth=2,markersize=5,label='S')
plt.xlabel("X")
plt.ylabel("Y")
plt.axis([-0.5,10.5,-5,105])
plt.legend(loc=2)
plt.savefig("mplot1.pdf");
```
